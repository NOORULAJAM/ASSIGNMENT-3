```python
math = int(input("ENTER MATH NUMBER  "));
physics = int(input(" ENTER PHYSICS NUMBER  "));
chemistry = int(input("ENTER CHEMISTRY NUMBER  "));
urdu = int(input(" ENTER URDU NUMBER  "));
science = int(input("ENTER SCIENCE NUMBER  "));

total= math+ physics+chemistry+urdu+science;
percentage = int(total/5);
if percentage >= 80:
    print("YOUR GRAGE A+1")
elif percentage >= 70:
    print("YOUR GRAGE A")
elif percentage >= 60:
    print("YOUR GRAGE B")
elif percentage >= 50:
    print("YOUR GRAGE C") 
elif percentage >= 40:
    print("YOUR GRAGE D")  
elif percentage >= 33:
    print("YOUR GRAGE E")
elif percentage < 33:
    print("FAIL");
else:
    print("ENTER CORRECT NUMBER");

    
print("YOUR PERCENTAGE " +str(percentage) +"%" );
print("YOUR GRAGE");
```
